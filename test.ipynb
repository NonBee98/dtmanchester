```python
from fastpip import pip
import matplotlib.pyplot as plt
```

```python
data = [(0, 0), (1, 0.5), (2, 1.8), (3, 1), (4, 1.2), (5, 0.7), (6, 1.6), (7, 1), (8, 0.5), (9, 0.9)]
```

```python
m = pip(data, 5)
```

```python
x = [i[0] for i in data]
y = [i[1] for i in data]
```

```python
plt.plot(x,y)
plt.scatter(x, y, color='b')
plt.scatter(x, y, color='', marker='o', edgecolors='g', s=200) 
plt.xticks([]) 
plt.yticks([])
plt.axis('off')
plt.savefig('original.jpg', dpi=200)
```

```python
a = [i[0] for i in m]
b = [i[1] for i in m]
```

```python
plt.plot(a,b)
plt.scatter(a, b, color='b')
plt.scatter(a, b, color='', marker='o', edgecolors='g', s=200) 
plt.xticks([]) 
plt.yticks([])
plt.axis('off')
plt.savefig('pip.jpg', dpi=200)
```

```python
from pyts.approximation import PiecewiseAggregateApproximation
```

```python
transformer = PiecewiseAggregateApproximation(window_size=2)
ppa = transformer.transform([x,y])
```

```python
a = ppa[0]
b = ppa[1]
plt.plot(a,b)
plt.scatter(a, b, color='b')
plt.scatter(a, b, color='', marker='o', edgecolors='g', s=200) 
plt.xticks([]) 
plt.yticks([])
plt.axis('off')
plt.savefig('ppa.jpg', dpi=200)
```
